## 3.3.2 applying data transformation

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, random_state=1)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
# printing the contributes of dataset after scaling
print("shape after scaling: ", X_train_scaled.shape)
print("Minimum of each feature before scaling:\n", X_train.min(axis=0))
print("Maximum of each feature before scaling:\n", X_train.max(axis=0))

print("Minimum of each feature after scaling:\n", X_train_scaled.min(axis=0))
print("Maximum of each feature after scaling:\n", X_train_scaled.max(axis=0))
```

```python
X_test_scaled = scaler.transform(X_test)

print("Minimum of each feature after scaling:\n", X_test_scaled.min(axis=0))
print("Maximum of each feature after scaling:\n", X_test_scaled.max(axis=0))
```

## 3.3.3 Quantile Transformer and PowerTransformer

```python
from sklearn.datasets import make_blobs
from sklearn.preprocessing import QuantileTransformer, StandardScaler, PowerTransformer
```

```python
X, y = make_blobs(n_samples=50, centers=2, random_state=4, cluster_std=1)
X += 3

plt.scatter(X[:,0], X[:, 1], c=y, s=30, edgecolors='black')
plt.xlim(0,16)
plt.xlabel('x0')
plt.ylim(0, 10)
plt.ylabel('x1')
plt.title("original Data")
plt.show()
```

```python
scaler = QuantileTransformer()
X_trans = scaler.fit_transform(X)

plt.scatter(X_trans[:,0], X_trans[:,1], c=y, s=30, edgecolors='black')
plt.xlim(0,5)
plt.xlabel('x0')
plt.ylim(0,5)
plt.ylabel('x1')
plt.title(type(scaler).__name__)
plt.show()
```

```python
plt.hist(X_trans)
plt.show()
```

```python
print(scaler.quantiles_.shape)
```

```python
x = np.array([[0], [5], [8], [9], [10]])
print(np.percentile(x[:,0], [0, 25, 50, 75, 100]))
```

```python
x_trans = QuantileTransformer().fit_transform(x)
print(np.percentile(x_trans[:,0], [0, 25, 50, 75, 100]))
```

```python
scaler = QuantileTransformer(output_distribution='normal')
X_trans = scaler.fit_transform(X)
plt.scatter(X_trans[:,0], X_trans[:,1], c=y, s=30, edgecolors='black')
plt.xlim(-5, 5)
plt.xlabel('x0')
plt.ylim(-5, 5)
plt.ylabel('x1')
plt.title(type(scaler).__name__)
plt.show()
```

```python
plt.hist(X)
plt.title('original Data')
plt.show()

X_trans = QuantileTransformer(output_distribution='normal').fit_transform(X)
plt.hist(X_trans)
plt.title('QuantileTransformer')
plt.show()

X_trans = StandardScaler().fit_transform(X)
plt.hist(X_trans)
plt.title('StandardScaler')
plt.show()

X_trans = PowerTransformer(method='box-cox').fit_transform(X)
plt.hist(X_trans)
plt.title('PowerTransformer')
plt.show()

X_trans = PowerTransformer(method='yeo-johnson').fit_transform(X)
plt.hist(X_trans)
plt.title('PowerTransformer, yeo-johnson')
plt.show()
```

## 3.3.4 scaling test data same as training data

```python
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=50, centers=5, random_state=4, cluster_std=2)

X_train, X_test = train_test_split(X, random_state=5, test_size=.1)

fig, axes = plt.subplots(1, 3, figsize=(13,4))
axes[0].scatter(X_train[:,0], X_train[:,1],
               c=mglearn.cm2.colors[0], label="train set", s=60)
axes[0].scatter(X_test[:,0], X_test[:,1], marker='^',
               c=mglearn.cm2.colors[1], label="test set", s=60)
axes[0].legend(loc='upper left')
axes[0].set_title("original data")

scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

axes[1].scatter(X_train_scaled[:,0], X_train_scaled[:,1],
               c=mglearn.cm2.colors[0], label="train set", s=60)
axes[1].scatter(X_test_scaled[:,0], X_test_scaled[:,1], marker='^',
               c=mglearn.cm2.colors[1], label="test set", s=60)
axes[1].set_title("scaled data")

test_scaler = MinMaxScaler()
test_scaler.fit(X_test)
X_test_scaled_badly = test_scaler.transform(X_test)

axes[2].scatter(X_train_scaled[:,0], X_train_scaled[:,1],
               c=mglearn.cm2.colors[0], label="train set", s=60)
axes[2].scatter(X_test_scaled_badly[:,0], X_test_scaled_badly[:,1], marker='^',
               c=mglearn.cm2.colors[1], label="test set", s=60)
axes[2].set_title("badly scaled data")

for ax in axes:
    ax.set_xlabel("feature 0")
    ax.set_ylabel("feature 1")
```

## 3.3.5 the effect of pre-proccessing in supervised learning

```python
from sklearn.svm import SVC

X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, random_state=0)

svm = SVC(C=100)
svm.fit(X_train, y_train)
print("test accuracy: ", svm.score(X_test, y_test))
```

```python
# scaling with MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# svm learning with scaled data
svm.fit(X_train_scaled, y_train)
print("scaled test set accuracy: ", svm.score(X_test_scaled, y_test))
```

```python
# scaling with StandardScaler (average=0, distribution=1)
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# svm learning with scaled data
svm.fit(X_train_scaled, y_train)
print("scaled test set accuracy: ", svm.score(X_test_scaled, y_test))
```
